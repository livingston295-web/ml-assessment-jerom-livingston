```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df = pd.read_csv('q2_customers.csv')

scaler = StandardScaler()
scaled = scaler.fit_transform(df)

# Elbow
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(scaled)
    wcss.append(km.inertia_)

plt.plot(range(1,11), wcss)
plt.show()

# KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled)

# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pcs = pca.fit_transform(scaled)

plt.scatter(pcs[:,0], pcs[:,1], c=df['cluster'])
plt.show()
```

Data is scaled because K-Means uses distance-based calculations.


Customers are grouped into clusters based on similar behavior.


The elbow method helps determine the optimal number of clusters.


PCA reduces data into two dimensions for visualization.
